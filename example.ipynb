```python
class two_sent_dataset(torch.utils.data.Dataset):
    def __init__(self, data_dict: dict):
        self.data_dict = data_dict
        dd = data_dict

        if len(dd["first"]) != len(dd["second"]) or len(dd["second"]) != len(
            dd["label"]
        ):
            raise AttributeError("Incorrect length in data_dict")

    def __len__(self):
        return len(self.data_dict["first"])

    def __getitem__(self, idx):
        dd = self.data_dict
        return dd["first"][idx], dd["second"][idx], dd["label"][idx]
```

```python
class AlbertModel(nn.Module):
    def __init__(self):
        super().__init__()
        self.tokenizer = transformers.AlbertTokenizer.from_pretrained('albert-xxlarge-v2')
        self.distilbert  =  transformers.AlbertModel.from_pretrained("albert-xxlarge-v2")
        self.pred_layer = nn.Linear(self.distilbert.config.hidden_size, 1)
        self.sigmoid = nn.Sigmoid()
        self.criterion = nn.BCELoss()
        self.n_tokens = 128
        self.initialize_from_vocab = True
        s_wte = SoftEmbedding(self.distilbert.get_input_embeddings(), 
                      n_tokens=self.n_tokens, 
                      initialize_from_vocab=self.initialize_from_vocab)
        self.distilbert.set_input_embeddings(s_wte)
        
        for param in self.distilbert.parameters():
            param.requires_grad = False
            
        for param in self.distilbert.get_input_embeddings().parameters():
            param.requires_grad = True
        
    def get_distilbert(self):
        return self.distilbert

    def get_tokenizer(self):
        return self.tokenizer

    def get_pred_layer(self):
        return self.pred_layer

    def get_sigmoid(self):
        return self.sigmoid
    
    def get_criterion(self):
        return self.criterion

    def assign_optimizer(self, **kwargs):
        model_params = self.parameters()
        optimizer = torch.optim.Adam(model_params, **kwargs)  
        return optimizer

    def slice_cls_hidden_state(
        self, x: transformers.modeling_outputs.BaseModelOutput
    ) -> torch.Tensor:
        last_states = x.last_hidden_state
        result =last_states[:, 0, :]
        return result

    def tokenize(
        self,
        premise: "list[str]",
        hypothesis: "list[str]",
        max_length: int = 128,
        truncation: bool = True,
        padding: bool = True,
    ):
        tokenizer = self.get_tokenizer()
        inputs = tokenizer(premise, hypothesis, padding=True, max_length=max_length, truncation=True, return_tensors="pt")
        #inputs['input_ids'] = torch.cat([torch.full((1, self.n_tokens), 50256), inputs['input_ids']], 1)
        #inputs['attention_mask'] = torch.cat([torch.full((1,self.n_tokens), 1), inputs['attention_mask']], 1)
        return inputs

    def forward(self, inputs: transformers.BatchEncoding):
        outputs = self.distilbert(**inputs)
        cls_token_hidden_state = self.slice_cls_hidden_state(outputs)
        zero_one = self.pred_layer(cls_token_hidden_state)
        predictions = self.sigmoid(zero_one)
        return predictions[:, 0]
```
